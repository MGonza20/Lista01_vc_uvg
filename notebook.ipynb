```python
# Librerias necesarias

import cv2 as cv
import numpy as np
import os

from matplotlib import pyplot as plt
from PIL import Image
```

## Inciso 1

### OpenCV

```python
# Leyendo imagen con OpenCV
img_opencv = cv.imread('imgs/saigon.jpeg', cv.IMREAD_COLOR)
img_opencv_rgb = cv.cvtColor(img_opencv, cv.COLOR_BGR2RGB) # convirtiendo de BGR a RGB

# Mostrando imagen
plt.imshow(img_opencv_rgb)
plt.axis('off')
plt.show()
```

```python
# Almacenando imagen - OpenCV
dir_exists = cv.imwrite('imgs/save/saigon_opencv.jpeg', img_opencv) # Almacenando imagen en formato JPG

if not dir_exists: 
    os.makedirs('imgs/save')
    cv.imwrite('imgs/save/saigon_opencv.jpeg', img_opencv)
    
```

```python
# Almacenando imagen con Matplotlib
plt.imsave('imgs/save/saigon_matplotlib.png', img_opencv_rgb)
```

```python
# Convertir a un array de numpy - OpenCV
img_opencv # En el caso de OpenCV, no es necesario porque la imagen se almacena en un array de numpy por defecto
```

La imagen se almacena en memoria como un arreglo tridimensional de NumPy. La primera dimensin corresponde a la altura de la imagen, reflejando la cantidad de filas. Por otra parte, la segunda dimensin indica el ancho de la imagen, donde cada elemento dentro de una fila equivale a un pxel, cada uno se representa por un arreglo unidimensional, representando a la tercera dimensin. En este arreglo, se almacena la informacin de color de cada pxel siguiendo el formato BGR (Azul, Verde, Rojo).
